### matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# Chinese Language support
# font1 = FontProperties(fname=r"c:\windows\fonts\simsun.ttc",size=20) 
# plt.title(u"标题",fontproperties=font1)

# add annoation
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2, height, height, ha='center', va='bottom')

# code for visualization
x1 = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6']
y1 = [5, 20, 36, 10, 75, 90]
y2 = [15, 25, 41, 5, 12, 32]
plt.figure(figsize = (12,8))
p1 = plt.bar(range(len(x1)), y1, width = 0.4, label = 'testy1')
p2 = plt.bar(np.arange(len(x1)) + 0.3, y2, width = 0.4, label = 'testy2')
plt.xticks(range(len(x1)), x1)
plt.title('Chart Test', fontsize = 16)
plt.legend(fontsize = 13)
plt.xlabel('Test_x')
plt.ylabel('Test_y')
# add annoation
add_labels(p1)
add_labels(p2)
plt.show()
```

### seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_style('darkgrid')
data = pd.DataFrame({'x': ['x1', 'x2', 'x3', 'x4', 'x5', 'x6']*2, \
                     'y': [5, 20, 36, 10, 75, 90, 15, 25, 41, 5, 12, 32], \
                     'tag': ['y1' for i in range(6)] + ['y2' for i in range(6)]})
sns.barplot(x = 'x', y = 'y', hue = 'tag', data = data)
plt.title('Chart Test', fontsize = 13)
plt.legend(fontsize = 13)
plt.xlabel('Test_x')
plt.ylabel('Test_y')
plt.show()
```

### plotly example

```python
import plotly as py
import plotly.graph_objs as go
pyplt = py.offline.plot
# add trace
trace1 = go.Bar(
    x = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6'],
    y = [5, 20, 36, 10, 75, 90],
    name = 'testy1'
)
trace2 = go.Bar(
    x = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6'],
    y = [15, 25, 41, 5, 12, 32],
    name = 'testy2'
)
# all traces to data
data = [trace1, trace2]
# set layout
layout = go.Layout(
    title = 'Chart Test',
    yaxis = dict(
        title = 'TestLegend',
    )
)
fig = dict(data = data, layout = layout)
pyplt(fig, filename = 'test.html')
```

### pyecharts example

```python
import pyecharts as ech
bar = ech.Bar(title = 'Chart Test', subtitle = 'subtitle test')
# is_more_utils means whether we need more tools
# is_label_show means whether show the label
# is_datazoom_show means whether show the zoom tool
bar.add('TestLegend1', ['x1', 'x2', 'x3', 'x4', 'x5', 'x6'], [5, 20, 36, 10, 75, 90], \
        is_more_utils = True, is_label_show = True, is_datazoom_show = True)
bar.add('TestLegend2', ['x1', 'x2', 'x3', 'x4', 'x5', 'x6'], [15, 25, 41, 5, 12, 32], \
        is_more_utils = True, is_label_show = True, is_datazoom_show = True)
bar.render(path = 'echarts.html')
bar
```
